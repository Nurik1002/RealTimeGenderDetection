```python
import tensorflow as tf
import numpy as np
import cvlib
import cv2
```

```python
model = tf.keras.models.load_model("model.h5")
label = ['female', 'male']
```

```python
def real_time():
    video = cv2.VideoCapture(0)
    while True:
        red, fram = video.read()
        print(red)


        faces = cvlib.detect_face(fram)
        #print(faces)

        for x1,y1,x2,y2 in faces[0]:

            face = fram[y1-10:y2+10, x1-10:x2+10] 
            face = cv2.cvtColor(face, cv2.COLOR_BGR2RGB)        
            face = cv2.resize(face, (224,224))
            face = np.expand_dims(face, 0)   

            pred = np.argmax(model.predict(face))

            if pred == 0:
                cv2.rectangle(fram, (x1,y1), (x2,y2), (0,0,255),2)
                cv2.putText(fram, label[pred].upper(), (x1,y1), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,255),2)

            else:
                cv2.rectangle(fram, (x1,y1), (x2,y2), (255,0,0),2)
                cv2.putText(fram, label[pred].upper(), (x1,y1), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,0,0),2)

        cv2.imshow("Video frame", fram)


        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

    video.release()
    cv2.destroyAllWindows()

```
