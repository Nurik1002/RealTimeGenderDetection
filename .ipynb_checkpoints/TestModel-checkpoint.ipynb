```python
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np
import cvlib
import cv2
```

```python
model = tf.keras.models.load_model("model.h5")
pred = ['female', 'male']
```

```python
img1 = cv2.imread("images/test1.jpg")
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img1_copy = img1.copy()
plt.imshow(img1)
plt.xticks([])
plt.yticks([])
plt.title("Test 1")
plt.show()
```

```python
face1 = cvlib.detect_face(img1)
for x1,y1,x2,y2 in face1[0]:
   
    img = img1[y1-10:y2+10, x1-10:x2+10]
    img = cv2.resize(img, (224,224))
    img = np.expand_dims(img, 0)
     
        
    natija = np.argmax(model.predict(img))
    if natija == 0:
        cv2.rectangle(img1_copy,(x1-10,y1-10),(x2+10,y2+10),(255,0,0),3)
    else :
        cv2.rectangle(img1_copy,(x1-10,y1-10),(x2+10,y2+10),(0,0,255),3)
        
    
plt.imshow(img1_copy)
plt.title("Red Female\nBlue Male")
plt.xticks([])
plt.yticks([])
plt.show()
```
